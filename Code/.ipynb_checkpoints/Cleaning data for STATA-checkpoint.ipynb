# Here in Python I will prepare the data for STATA usage

Import pandas library and read xls file for future corrections

```python
import numpy as np
import pandas as pd
data = pd.read_excel('../Data/Raw/online_offline_ALL_clean.xls') 
```

```python
#let's see how it looks
data.head()
```

```python
def replacenan(ind):
    if pd.isna(data.iloc[ind,12]): #PRICETYPE is 12th column by index
        if data.iloc[ind,11]=='s': #check if COMMENTS has 's'
            data.iloc[ind,12]='Sale/Discounted Price'
        else:
            data.iloc[ind,12]='Regular Price'
```

```python
for ind in data.index:
    replacenan(ind)
```
